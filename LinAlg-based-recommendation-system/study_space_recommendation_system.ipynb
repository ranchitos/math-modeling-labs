```python
import numpy as np
```

```python
# study spaces
spaces = {
    "library": np.array([1,2]),
    "cafe": np.array([4,4]),
    "lounge": np.array([3,2]),
}
spaces
```

```python
# student preferences
studi = np.array([2,1])
```

```python
# compute distances
distances = {}
for name, vec in spaces.items():
    distances[name] = np.linalg.norm(studi - vec)

distances
```

```python
# recommend the closest space
min_distance = min(distances.values())
best_space = [name for name, distance in distances.items() if np.isclose(distance, min_distance)]
best_space
```

```python
# we can also do the same thing with KNN algorithm which is what we literally did above
from sklearn.neighbors import NearestNeighbors
import numpy as np

X = np.array(list(spaces.values()))
labels = np.array(list(spaces.keys()))

model = NearestNeighbors(n_neighbors=2, metric="euclidean")
model.fit(X)

distances, indices = model.kneighbors([studi])

recommended = labels[indices[0]]
print(recommended)
```
